```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import random
import time
import tqdm
```

```python
total_pags = 1317 # - apartamentos
#total_pags = 901 # - casas
start_url = "https://www.wimoveis.com.br/apartamentos-venda-distrito-federal-goias.html"
proto_url = "https://www.wimoveis.com.br/apartamentos-venda-distrito-federal-goias-pagina-{}.html"
#start_url = "https://www.wimoveis.com.br/casas-venda-distrito-federal-goias.html"
#proto_url = "https://www.wimoveis.com.br/casas-venda-distrito-federal-goias-pagina-{}.html"
print(proto_url.format(2))
```

```python
response = requests.get(start_url)

```

```python
soup = BeautifulSoup(response.text, 'html.parser')
titulos_pre = soup.findAll("a",{"class":"go-to-posting"})
valores_pre = soup.findAll("span", {"class": "first-price"},text=True)
```

```python
with open('test.html', 'w') as outfile:
    outfile.write(response.text)
```

```python
urls = []
titulos = []
for titulo in titulos_pre:
    urls.append("https://www.wimoveis.com.br" + titulo['href'])
    titulos.append(titulo.text.replace("\t","").replace("\n",""))
```

```python
valores = []
for valor in valores_pre:
    valores.append(valor.text.replace("R$ ","").replace(".",""))
```

```python
len(titulos),len(valores),len(urls)
```

```python
#for x,y,z in zip(titulos, valores, urls):
#    print(x,y,z)
```

```python
df = pd.DataFrame(list(zip(titulos, valores, urls)),
              columns=['nome','valor_venda', 'url'])
```

```python
print(df.shape)
df.head()
```

```python
for num in tqdm.tqdm_notebook(range(2,total_pags), total=len(range(2,total_pags))):

#for num in range(2,total_pags):
    try:
        print(num)
        print("pegando pagina",proto_url.format(num))
        response = requests.get(proto_url.format(num))
        soup = BeautifulSoup(response.text, 'html.parser')
        titulos_pre = soup.findAll("a",{"class":"go-to-posting"})
        valores_pre = soup.findAll("span", {"class": "first-price"},text=True)
        urls = []
        titulos = []
        valores = []

        for titulo in titulos_pre:
            urls.append("https://www.wimoveis.com.br" + titulo['href'])
            titulos.append(titulo.text.replace("\t","").replace("\n",""))

        for valor in valores_pre:
            valores.append(valor.text.replace("R$ ","").replace(".",""))

        df_pag = pd.DataFrame(list(zip(titulos, valores, urls)),
                  columns=['nome','valor_venda', 'url'])

        df = df.append(df_pag)

    except:
        pass
        
    #dorme
    aleatorio = random.randint(1, 5)
    print("agora vou aguardar:",aleatorio)
    time.sleep(aleatorio)
```

```python
print(df.shape)
df['downloaded'] = False
df.head()
```
